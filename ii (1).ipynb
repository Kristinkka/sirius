```python
import json
import requests
```

```python
def get_next_page(page):
    headers = {
        'authority': 'ms-gateway.tinkoff.ru',
        'accept': '*/*',
        'accept-language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
        'content-type': 'application/json',
        'origin': 'https://www.tinkoff.ru',
        'referer': 'https://www.tinkoff.ru/',
        'sec-ch-ua': '"Chromium";v="122", "Not(A:Brand";v="24", "Google Chrome";v="122"',
        'sec-ch-ua-mobile': '?0',
        'sec-ch-ua-platform': '"Windows"',
        'sec-fetch-dest': 'empty',
        'sec-fetch-mode': 'cors',
        'sec-fetch-site': 'same-site',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
    }
    params = {
        'nextPage': page,
        'sort[0]': 'id,desc',
        'sort[1]': 'createdDate,desc',
        'size': '100',
    }
    json_data = {
        'onlyText': True,
        'ratings': [],
        'bmsBrandId': 102402,
    }
    response = requests.post(
        "https://ms-gateway.tinkoff.ru/city/public/v2/feedback/search",
        params=params,
        headers=headers,
        json=json_data,
    )
    data = response.json()
    return [data["nextPage"], data["content"]]
```

```python
first_page = "ZmVlZGJhY2tJZD0zMjk3NjYwMDtjcmVhdGVkRGF0ZT0xNzExNzAzMDQ4NTIxO3Njcm9sbGVkT3V0RmVlZGJhY2tzPTYwOw=="
res = []
```

```python
next_page = first_page
```

```python
for i in range(100):
    temp = get_next_page(next_page)
    next_page = temp[0]
    for review in temp[1]:
        res.append([review["text"], review["rating"]])
```
